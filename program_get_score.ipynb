# Buka File di Direktori

```python
import pandas as pd
import numpy as np

# Database (File Excel) untuk menyimpan nilai final
dataset =  pd.read_excel('2023_D4TRII_daftar_mahasiswa.xlsx')

# Daftar nilai
nilai = pd.read_csv('Angkatan 2023\Pendidikan_Agama_Softskill.csv')
```

# Membersihkan data

```python
# Ubah semua huruf alamat email menjadi huruf kecil
nilai = nilai.rename(columns={'Email Address': 'Alamat Email'}) # Ubah nama header 'Email Address' menjadi 'Alamat Email' jika perlu
dataset['Email'] = dataset['Email'].str.lower()
nilai['Alamat Email'] = nilai['Alamat Email'].str.lower()

# mendapatkan daftar header nilai
nilai_head = list(nilai.columns)
dataset_head = list(dataset.columns)

# tambahkan kolom kosong baru seperti nilai absensi, tugas dll
dataset = dataset.reindex(columns = dataset.columns.tolist() + nilai_head[3:])
dataset
```

# Mencocokkan Data

```python
# yang berisi NaN tidak perlu dibandingkan
periksa_dataset = dataset['Email'].isnull()

for i in range(0,len(dataset)):
    if(~periksa_dataset[i]):
        for j in range(0,len(nilai)):
            if(dataset['Email'][i]==nilai['Alamat Email'][j]):
                for k in range(3,len(nilai_head)):
                    dataset[nilai_head[k]][i] = nilai[nilai_head[k]][j]
```

```python
dataset
```

```python
dataset = dataset.reindex(sorted(dataset.columns), axis=1)

first_cols = dataset_head[:4]
last_cols = [col for col in dataset.columns if col not in first_cols]

dataset = dataset[first_cols+last_cols]
```

```python
dataset.to_excel("output.xlsx")

# writer = pd.ExcelWriter('output.xlsx',
#                         engine='xlsxwriter',
#                         options={'strings_to_numbers': True})
```
