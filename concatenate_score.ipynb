# 1. Buka File di Direktori
Defenisikan lokasi data dua output dari <i>program_get_score.ipynb</i> yang ingin digabungkan

```python
import pandas as pd
import numpy as np

# Database Nilai Pertama
data1 =  pd.read_excel('output1.xlsx')
data1.drop(data1.columns[0], axis=1, inplace=True)

# Database Nilai Kedua
data2 = pd.read_excel('output2.xlsx')
data2.drop(data2.columns[0], axis=1, inplace=True)
```

# Membersihkan data

```python
# mendapatkan daftar header
dataset_head = list(data1.columns)
```

```python
gabung = data1.combine_first(data2)
```

```python
gabung = gabung.reindex(sorted(gabung.columns), axis=1)

first_cols = dataset_head[:4]
last_cols = [col for col in gabung.columns if col not in first_cols]

gabung = gabung[first_cols+last_cols]
```

```python
gabung.to_excel("output_gabung.xlsx")

# writer = pd.ExcelWriter('output.xlsx',
#                         engine='xlsxwriter',
#                         options={'strings_to_numbers': True})
```
